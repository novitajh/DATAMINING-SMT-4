```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt

seattle_weather = pd.read_csv("seattle-weather.csv")
seattle_weather.head()

numerical_columns = ['precipitation', 'temp_max', 'temp_min', 'wind']
scaler = MinMaxScaler()
seattle_weather[numerical_columns] = scaler.fit_transform(seattle_weather[numerical_columns])
seattle_weather.head()
```

4. Tentukan K terbaik jika dilakukan dengan analisis Elbow Method



```python
attribute_combinations = [('precipitation', 'temp_max'),
                          ('temp_min', 'wind'),
                          ('precipitation', 'temp_min'),
                          ('temp_max', 'wind')]
optimal_K = {}

for i, (attribute1, attribute2) in enumerate(attribute_combinations, start=1):
    X = seattle_weather[[attribute1, attribute2]]

    inertia = []
    K_range = range(1, 11)
    for k in K_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(X)
        inertia.append(kmeans.inertia_)

    diff = [inertia[j] - inertia[j-1] for j in range(1, len(inertia))]
    optimal_K[i] = diff.index(max(diff)) + 2

    plt.figure(figsize=(8, 5))
    plt.plot(K_range, inertia, marker='o')
    plt.title('Elbow Method for {} - {}'.format(attribute1, attribute2))
    plt.xlabel('Number of Clusters (K)')
    plt.ylabel('Inertia')
    plt.xticks(K_range)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

```

5. Tampilkan scatter plot setelah clustering

```python
attribute_combinations = [('precipitation', 'temp_max'),
                         ('temp_min', 'wind'),
                         ('precipitation', 'temp_min'),
                         ('temp_max', 'wind')]

for i, (attribute1, attribute2) in enumerate(attribute_combinations, start=1):
   X = seattle_weather[[attribute1, attribute2]]

   k_values = [2, 3, 4]

   for k in k_values:
       kmeans = KMeans(n_clusters=k, random_state=42)
       seattle_weather['cluster_' + str(k)] = kmeans.fit_predict(X)

       plt.figure(figsize=(6, 4))
       plt.scatter(X[attribute1], X[attribute2], c=seattle_weather['cluster_' + str(k)], cmap='viridis', alpha=0.5)
       plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=100, c='red', label='Centroids')
       plt.title('Scatter Plot Setelah Clustering (K = {}) - {}, {}'.format(k, attribute1, attribute2))
       plt.xlabel(attribute1)
       plt.ylabel(attribute2)
       plt.legend()
       plt.tight_layout()
       plt.show()
```
