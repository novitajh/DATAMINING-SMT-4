1. Masukkan iris.csv ke dataframe. Normalisasikan setiap atribut iris ke dataframe menggunakan normalisasi Min-Max sehingga setiap nilai atribut memiliki range antara 0 sampai 1. Dengan Vbaru adalah nilai data disuatu sel setelah normalisasi, Vlama adalah nilai data asli sebelum normalisasi, Vmax adalah nilai max dari kolom data tersebut, dan Vmin adalah nilai min dari kolom data tersebut.


```python
import pandas as pd
df_iris = pd.read_csv('iris.csv')
df_iris.head()
```

```python
# Membaca file iris.csv ke dalam dataframe
df_iris = pd.read_csv('iris.csv')

# Fungsi untuk melakukan normalisasi Min-Max
def min_max_normalization(x):
    return (x - x.min()) / (x.max() - x.min())

# Kolom yang akan dinormalisasi
columns_to_normalize = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

# Melakukan normalisasi Min-Max untuk setiap kolom
df_normalized = df_iris.copy()
df_normalized[columns_to_normalize] = df_normalized[columns_to_normalize].apply(min_max_normalization)

# Menampilkan dataframe yang sudah dinormalisasi beserta Vbaru, Vlama, Vmax, dan Vmin
for col in columns_to_normalize:
    Vbaru = df_normalized[col].values
    Vlama = df_iris[col].values
    Vmax = df_iris[col].max()
    Vmin = df_iris[col].min()
    print(f'Kolom: {col}')
    print('Vbaru:', Vbaru[:5])
    print('Vlama:', Vlama[:5])
    print('Vmax:', Vmax)
    print('Vmin:', Vmin)
    print()

# Tampilkan dataframe yang sudah dinormalisasi
print(df_normalized.to_string(index=False))
```

2. Carilah centroid dari data yang telah dinormalisasi. Centroid dihitung dari rata-rata seluruh data suatu kolom.



```python
# Cari centroid (rata-rata) dari data yang telah dinormalisasi
centroid = df_normalized.mean()

# Tampilkan centroid
print("Centroid:")
print(centroid)
```

3. Masukkan iristesting.csv ke dataframe. Normalisasikan iristesting.csv menggunakan normalisasi Min-Max.

```python
import pandas as pd
df_iristesting = pd.read_csv('iristesting.csv')
print(df_iristesting.to_string())
```

```python
sepal_length_max = df_iristesting['sepal.length'].max()
sepal_length_min = df_iristesting['sepal.length'].min()
sepal_width_max = df_iristesting['sepal.width'].max()
sepal_width_min = df_iristesting['sepal.width'].min()
petal_length_max = df_iristesting['petal.length'].max()
petal_length_min = df_iristesting['petal.length'].min()
petal_width_max = df_iristesting['petal.width'].max()
petal_width_min = df_iristesting['petal.width'].min()

for i in df_iristesting.index :
  df_iristesting.loc[i, 'sepal.length'] = (df_iristesting.loc[i, 'sepal.length']-sepal_length_min)/(sepal_length_max-sepal_length_min)
  df_iristesting.loc[i, 'sepal.width'] = (df_iristesting.loc[i, 'sepal.width']-sepal_width_min)/(sepal_width_max-sepal_width_min)
  df_iristesting.loc[i, 'petal.length'] = (df_iristesting.loc[i, 'petal.length']-petal_length_min)/(petal_length_max-petal_length_min)
  df_iristesting.loc[i, 'petal.width'] = (df_iristesting.loc[i, 'petal.width']-petal_width_min)/(petal_width_max-petal_width_min)

print(df_iristesting.to_string())
```

4. Prediksilah kelas data di iristesting.csv menggunakan kriteria kedekatan distance antara setiap data di iristesting.csv ke centroid masing2 class di iris. Gunakan euclidean distance.

```python
import pandas as pd
def euclidean(data, dataframe):
    classes = ['setosa', 'versicolor', 'virginica']
    distances = []
    result = None

    for cls in classes:
        class_data = dataframe[dataframe['variety'] == cls]
        centroid = class_data.mean()
        test = dataframe.iloc[data, 0:4]
        distance = ((centroid[0] - test[0]) ** 2 + (centroid[1] - test[1]) ** 2 +
                    (centroid[2] - test[2]) ** 2 + (centroid[3] - test[3]) ** 2) ** 0.5
        distances.append(distance)

    if min(distances) == distances[0]:
        print("Setosa True")
        result = 'Setosa'
    elif min(distances) == distances[1]:
        print("Versicolor True")
        result = 'Versicolor'
    else:
        print("Virginica True")
        result = 'Virginica'

    return result

data = []
for i in df_iristesting.index:
    data.append(euclidean(i, df_iristesting))
```

5. Hitunglah berapa persen klasifikasi yang benar (akurasi) saat memprediksi class


```python
import pandas as pd

correct_predictions = sum(df_iristesting['variety'] == data)
total_data = len(df_iristesting)

accuracy = (correct_predictions / total_data) * 100
print(f'Akurasi: {accuracy:.2f}%')
```
